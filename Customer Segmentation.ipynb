# Applying K-Means Clustering

## Finding the number of clusters (i.e. value of K for K-Means Clustering) in our data for "Age" v/s "Spending Score (1-100)":

```python
X1=df.loc[:, ["Age","Spending Score (1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2, color="red", marker ="8")
plt.xlabel("k Value")
plt.ylabel("WCSS")
plt.show()
    
```

### Showing the cluster of each customers 

```python
kmeans = KMeans(n_clusters=4)

label = kmeans.fit_predict(X1)

print(label)
```

### Checking the centers of each cluster

```python
print(kmeans.cluster_centers_)
```

### Visualisation of four Clusters "

```python
plt.scatter(X1[:,0],X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color = 'black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

## Finding the number of clusters (i.e. value of K for K-Means Clustering) in our data for "Annual Income (k$)" v/s "Spending Score (1-100)":

```python
X2=df.loc[:, ["Annual Income (k$)","Spending Score (1-100)"]].values
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2, color="red", marker ="8")
plt.xlabel("k Value")
plt.ylabel("WCSS")
plt.show()
```

### After k Value=5 graph doesn't seem to be varying much,so K==5.

### Showing the cluster of each group

```python
kmeans = KMeans(n_clusters=5)

label = kmeans.fit_predict(X2)

print(label)
```

### Showing the center of each cluster

```python
print(kmeans.cluster_centers_)
```

### Visualization of five clusters 

```python
plt.scatter(X2[:,0],X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color = 'black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()
```

## Finding the number of clusters (i.e. value of K for K-Means Clustering) in our data for "Annual Income (k$)" v/s "Spending Score (1-100)" v/s "Age":

```python
X3=df.iloc[:,1:]
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2, color="red", marker ="8")
plt.xlabel("k Value")
plt.ylabel("WCSS")
plt.show()
```

### After value of k Value=5 , graph doesn't seem to be varying much, so K==5

### Showing the cluster of each group

```python
kmeans = KMeans(n_clusters = 5)


label = kmeans.fit_predict(X3)

print(label)
```

### Showing center of each cluster

```python
print (kmeans.cluster_centers_)
```

### visualization of all five clusters

```python
clusters = kmeans.fit_predict(X3)
df["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label == 0], df["Annual Income (k$)"][df.label == 0], df["Spending Score (1-100)"][df.label == 0], c='blue', s=60)
ax.scatter(df.Age[df.label == 1], df["Annual Income (k$)"][df.label == 1], df["Spending Score (1-100)"][df.label == 1], c='red', s=60)
ax.scatter(df.Age[df.label == 2], df["Annual Income (k$)"][df.label == 2], df["Spending Score (1-100)"][df.label == 2], c='yellow', s=60)
ax.scatter(df.Age[df.label == 3], df["Annual Income (k$)"][df.label == 3], df["Spending Score (1-100)"][df.label == 3], c='pink', s=60)
ax.scatter(df.Age[df.label == 4], df["Annual Income (k$)"][df.label == 4], df["Spending Score (1-100)"][df.label == 4], c='green', s=60)
ax.view_init(30, 185)

plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel("Spending Score (1-100)")

plt.show()

```
